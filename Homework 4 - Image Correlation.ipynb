##Homework 4: Image Correlation

For this homework, I intend to illustrate the use of image correlation as a technique for quality testing in manufacturing processes. If a camera is placed at a specific location in the assembly line, and if the objects arrive at the same exact location and orientation in the image, bending defects can be detected using correlation. 

To do so, I've taken the image of a pencil, and a second image of a pencil that is slightly bent (hardly noticable in plane sight). By correlating the two images, one should be able to visually see artefacts that show the stress mapping. 

This was partly inspired by: http://www.correlatedsolutions.com/digital-image-correlation/

####Preamble

```python
import numpy as np
import scipy.signal
import time
from matplotlib.pyplot import *
```

####Import the images:

```python
im1 = imread('Im1.jpg')
im2 = imread('Im2.jpg')
```

```python
imshow(im1)
show()
```

```python
imshow(im2)
show()
```

####Convert to Grayscale and Correlate

```python
# Convert to Grayscale by averaging the RGB channels
image1 = (im1[:,:,0] + im1[:,:,1] + im1[:,:,2])/3.0
image2 = (im2[:,:,0] + im2[:,:,1] + im2[:,:,2])/3.0
```

```python
# Resize the image (correlation taxes system for large images)
image1 = scipy.misc.imresize(image1, 0.25)
image2 = scipy.misc.imresize(image2, 0.25)
```

```python
t1 = time.time()
c = scipy.signal.correlate2d(image1, image2)
t2 = time.time()
print "Time Taken: %d", t2-t1
```

```python
imshow(c)
colorbar()
```
