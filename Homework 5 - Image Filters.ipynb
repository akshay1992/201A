####Helper Functions

Here are the definitions of some helper functions used throughout this code.

```python
def to_gray(image):
    R = image[:,:,0]
    G = image[:,:,1]
    B = image[:,:,2]
    return 0.299 * R + 0.587 * G + 0.114 * B 
```

```python
def binary_invert(image):
    return np.abs( image - np.ones(image.shape) )
```

```python
def expand(pixel, center, factor):
    centered = float(pixel) - center
    out = (centered*factor) + center
    if out > 255:
        out = 255
    if out < 0:
        out = 0
    return uint8(out)
    
vfunc = vectorize(expand)
```

```python
def kern_gauss(N):
    N = 20
    single_dot = zeros((N, N))
    single_dot[N/2.0, N/2.0] = 1.0
    gauss_kernel = ndimage.gaussian_filter(single_dot, N/4)
    return gauss_kernel
```

```python
def to_bin(image, thresh = -1, invert = False):
    if thresh == -1:
        thresh = np.mean(image)


    threshold = np.ones(image.shape) * thresh

    if invert:
        return where(np.greater_equal(threshold, image), 1, 0)
    else:
        return where(np.greater_equal(image, threshold), 1, 0)
```

```python
def dilateN(image, N):
    #N = Number of consecutive dialations
    
    structure = zeros((20,20))
    structure[range(20), range(20)] = 1.0

    for i in range(N):
        if i==0:
            dilated = ndimage.binary_dilation(im_edges, structure)
        else:
            dilated = ndimage.binary_dilation(dilated, structure)
    return dilated
```

##Main
The following process is carried out on the image:

* LPF
* Convert to Grayscale
* Dynamic Expansion (Excessive)
* Convert to Binary
* Apply morphological gradient
* Dilation: To yield two separate images
    * Dilation (x1 pass) 
    * Dilation (x2 passes)

And finally:

* Add the above two dilated pictures to the expanded image

```python
im = imread('coffee_killer.tiff')
imshow(im)
show()
```

Low-pass filtering the image to get the main figure.

```python
# Make the filtering kernel
kernel = kern_gauss(50)
l, b = kernel.shape
kern = np.zeros([l, b, 4])
kern[:,:,0] = kernel
kern[:,:,1] = kernel
kern[:,:,2] = kernel
```

```python
im_filtered = ndimage.convolve(im, kern) 
```

```python
# Convert to grayscale. 
im_gray = to_gray(im_filtered)
imshow(im_gray, cmap='gray')
colorbar()
```

```python
# Excessive dynamic expansion to get maximum contrast. 
expanded = vfunc(im_gray, 128, 10)
imshow(expanded, cmap=cm.gray)
```

```python
# Converting to binary to apply morphological gradient.
im_bin = to_bin(expanded)
imshow(im_bin, cmap='gray')
```

```python
# Trying to detect edges using morphological gradients. 
im_edges = morphology.morphological_gradient(im_bin, (10,10))
imshow(im_edges, cmap=cm.gray)
```

```python
# Multiple dilations on this edge line. 
dilate1 = dilateN(im_edges, 1)
dilate2 = dilateN(im_edges, 2)

subplot(1, 2, 1), imshow(dilate1, cmap='gray')
subplot(1, 2, 2), imshow(dilate2, cmap='gray')
```

```python
# Invert the dilated images to make the edge-areas white
dilate1 = binary_invert(dilate1)
dilate2 = binary_invert(dilate2)

subplot(1, 2, 1), imshow(dilate1, cmap='gray')
subplot(1, 2, 2), imshow(dilate2, cmap='gray')
```
