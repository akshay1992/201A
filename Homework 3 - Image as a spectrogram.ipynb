```python
# Resynthesize
def make_audio(spectrogram):
    NFFT,slices = spectrogram.shape
    
    
    sound = fft.ifft(spectrogram[:,0], n=NFFT)
    for window in range(1, slices):
        fftSlice = spectrogram[:, window]
        sound_slice = fft.ifft(fftSlice, n=NFFT)        
        sound = r_[sound, sound_slice]
    
    return sound

resynth = make_audio(sgram)
```
