This part does the following: 
* segments the audio into slices of length `win_len`. 
* Performs an FFT on each slice.
* Arranges each of these FFT-d slices as columns in a matrix

At the end of this, you will get a matrix that is effectively the spectrogram of the audio. 

```python
#Make a spectrogram
def segment_audio(audioFile, win_len):
    # Inits
    audio = audioFile.audio.copy()
    L = audioFile.length
    NFFT = win_len
    nslices = int( ceil(L*1.0/win_len) )
    
    # Generate Spectrogram
    spectrogram = fft.fft(audio[:win_len], n=NFFT)
    for w_index in range(1, nslices):
        # Make slice of audio
        window = audio[w_index*win_len : (1 + w_index)*win_len] 

        # Take FFT
        stft = fft.fft(window, n=NFFT)

        # Add to the matrix
        spectrogram = c_[spectrogram, stft]       
        
    return spectrogram
```

This is the algorithm that resynthesizes the audio from the spectrogram. The steps taken here are the exact inverse of those taken to make the spectrogram. 

* Take each column from the matrix. 
* Compute the inverse-FFT of this slice
* Append it to a long time-series (audio)

```python
# Resynthesize
def make_audio(spectrogram):
    NFFT,slices = spectrogram.shape
    
    # First slice
    sound = fft.ifft(spectrogram[:,0], n=NFFT)
    
    # Rest of the slices
    for window in range(1, slices):
        fftSlice = spectrogram[:, window]
        sound_slice = fft.ifft(fftSlice, n=NFFT)        
        sound = r_[sound, sound_slice]
    
    return sound
```

####Part 2: Adding overlapping to the algorithm

This is to obtain more time resolution in the spectrum thereby causeing the resynthesis to be less quantized (at least, in theory). This segment audio is almost the same as before. The only difference being that the part that makes the slice is slightly modified. 

The parameter `hop_size` determines by how many samples the window must 'hop' in order to make the next slice. Having a default value of -1 means that it hops without overlap (hops by one window length).

```python
def segment_audio(audioFile, win_len, hop_size=-1):    
    # Inits
    audio = audioFile.audio.copy()
    L = audioFile.length
    NFFT = win_len   
    if hop_size < 0:
        hop_size = win_len
        
    nslices = int( (L-NFFT)*1.0/hop_size )    
   
    # Generate Spectrogram
    spectrogram = fft.fft(audio[:win_len], n=NFFT)
    for w_index in range(1, nslices):
        # Make slice of audio
        window = audio[w_index*hop_size : w_index*hop_size + win_len]
        
        # Compute FFT of slice
        stft = fft.fft(window, n=NFFT)    
        
        # Add to the matrix
        spectrogram = c_[spectrogram, stft]       
        
    return spectrogram
```

The resynthesis part contains the following steps:

* Perform an inverse-FFT on each column.
* Add the last `hop_size` number of elements of one slice to the first `hop_size` elements of the subsequent slice.
* Get rid of the last `hop_size` elements of all columns (since you've added them already).
* Flatten it to make it 1D. 

```python
# Resynthesize
def make_audio(spectrogram, hop_size = -1, scale_level=1):    
    spgram = spectrogram.copy()
    NFFT,slices = spgram.shape
    
    if hop_size < 0:
        hop_size = NFFT

    # Perform ifft
    for i in range(slices):
        spgram[:,i] = fft.ifft(spgram[:,i], n=NFFT)

    # Overlap-add     
    head = spgram[:hop_size, :]
    tail = spgram[-hop_size:, :]
    head += tail
    
    # Make it 1D
    sound = (spgram[:hop_size, :]).copy()
    sound = sound.T.flatten()
            
    return sound

resynth = make_audio(sgram, hop_size=512)
```

###Part 3: Dealing with audio as a complex time-series. 

One final consideration that needs to be attended to is the fact that when dealing with images, the result of the inverse-FFT will not always be real. 

I have decided to deal with this by treating the real part and imaginary part as two separate signals and using them to create stereo. 

```python
# Resynthesize
def make_audio(spectrogram, hop_size = -1, scale_level=1):    
    spgram = spectrogram.copy()
    NFFT,slices = spgram.shape
    
    if hop_size < 0:
        hop_size = NFFT

    # Perform ifft
    for i in range(slices):
        spgram[:,i] = fft.ifft(spgram[:,i], n=NFFT)

    # Overlap-add     
    head = spgram[:hop_size, :]
    tail = spgram[-hop_size:, :]
    head += tail
    
    # Make it 1D
    sound = (spgram[:hop_size, :]).copy()
    sound = sound.T.flatten()
        
    # Make it a 2 channel sound. real - left, imaginary = right
    # This gives a very weird and interesting stereo effect. 
    R = real(sound)
    I = imag(sound)
    
    # Scale amplitude
    R = R * scale_level / np.max(abs(R))
    I = I * scale_level / np.max(abs(I))
    
    # Make Stereo
    output = c_[R, I]
    
    return output
```

I will also define a tiny helper function to plot stereo audio. 

```python
def plot_stereo(sound):    
    subplot(2, 1, 1), plot(sound[:, 0])
    title('Left channel')
    subplot(2, 1, 2), plot(sound[:, 1])
    title('Right channel')
```

In this final part I will import images and separate the RGB components using the following helper function. Using some complex combination of this RGB data, audio will be generated using the funciton defined in the previous part. 

```python
# Import image files
amazon = imread('amazon.jpg')
olympic = imread('Olympic-logo.png')
coke = imread('CokeLogo1.png')
```

Define the helper function. 

```python
def extractRGB(image):    
    R = image[:,:,0]
    G = image[:,:,1]
    B = image[:,:,2]
    return R, G, B
```
