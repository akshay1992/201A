###Converting an audio file into an image.

The audio is taken from a very short, nearly unrecognizable part of the song 'Hessian Peel' by Opeth. It is then chopped up into three pieces. Following this it uses a circular buffer to create a pattern using the samples.

Take three pieces in succession. Each for one color channels.

```python
size = 512
piece1 = a.audio[0:size]
piece2 = a.audio[size:2*size]
piece3 = a.audio[2*size:3*size] 
```

This takes a seed piece of audio and constructs a matrix by doing a circular-buffer type of shift. 

```python
def make_picture(seed):
    picture = seed.view()
    buff = r_[seed, seed]
   
    for i in range(len(seed)):
        new_addition = buff[i:i+len(seed)]
        picture = c_[picture, new_addition]

    return picture
```

Makes the picture by taking the three channels and constructing the three dimensional array

```python
# Do the circular buffering
r = make_picture(piece1)
g = make_picture(piece2)
b = make_picture(piece3)

# Assembling
picture = np.zeros([512, 513, 3])
picture[:,:,0] = r
picture[:,:,1] = g
picture[:,:,2] = b

#Rescaling
picture  = abs( picture / np.max(picture) )
```
