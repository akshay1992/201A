```python
def generateSpecgram(audioFile, windowFn = np.hanning, NFFT=1024, hop_size=-1):
    """ Generates a spectrogram of audio. Input audio must be in the form of an audioFile class instance """
    
    #Inits
    audio = audioFile.audio.copy()
    L = audioFile.length
    if hop_size < 0:
        hop_size = NFFT
        
    nslices = int( (L-NFFT)*1.0/hop_size )
    
    window = windowFn(NFFT)

    time_slice = audio[0 : NFFT]                      
    analysis_window = np.multiply(time_slice, window)
    stft = fft.fft(analysis_window, n=NFFT)    
    spectrogram = stft       

    for w_index in range(1, nslices):
        time_slice = audio[w_index*hop_size : w_index*hop_size + NFFT]                      
        analysis_window = np.multiply(time_slice, window)
        stft = fft.fft(analysis_window, n=NFFT)    
        spectrogram = c_[spectrogram, stft]       
        
    return spectrogram    
```

```python
def half_wave_rectify(array):
    # Returns 0 if negative, returns number if positive.
    # (Vectorized)
    return 0.5*(array+np.abs(array))
```

##Step : Onset detection

This method of onset detection uses the $L_2$ Norm on the rectified difference in each spectral bin as described in [1] and some improvements on this method as suggested by [2].

It involves the following steps:

* Compute the STFT with N=1024, Hanning, Hop Size=256
* Compute the difference in the magnitude of each bin with that of the previous analysis frame.
* Rectify the difference values using the following function to identify positive change.
    
    $H(x) = (x + |x|)/2 $ 
    
    This evaluates to zero for all negative values of $x$.
* Accumulate all these rectified values over each window. This yields a 1D signal called the Onset Strength Signal (OSS)

-----

[1] A Tutorial on Onset Detection in Music Signals, J. P. Bello L. Daudet, S Abdallah C Duxbury M Davies M B.Sandler

[2] Michael Krzyzaniak, ASU, mkrzyzan@asu.edu

The onset detection function for the above discussed procedure. 

```python
def detectOnsets(audioFile, NFFT=1024, hop=256, draw_plots=False):
    spgram = generateSpecgram(audioFile, np.hanning, NFFT, hop)
    specDiff = np.diff(abs(spgram))
    rectified = half_wave_rectify(specDiff)
    accum = np.sum(rectified, axis=0)
    
    # Draw plots to illustrate algorithm (if required)
    if(draw_plots):
        subplot(2,1,1), plot(audioFile.audio, label='Audio Signal'), legend()
        subplot(2,1,2), plot(accum, label='OSS'), legend()
        
    return accum

oss = detectOnsets(a, draw_plots=True)
```

##Step : Peak Picking

This is used to extract the exact locations of the peaks in the audio signal using a peak picking algorithm as described below. The following series of steps will attempt to illustrate the algorithm. This more or less follows some of the suggestions given in [1].

Algorithm:

* Smooth the OSS with a Hanning window. `(Window length: 64-points)`
* Make an adaptive threshold signal using a median filter on the smoothed-OSS. `(Kernel-size: 55)`
* Subtract this from the smoothed-OSS.
* Half-wave-rectify to isolate the signal that is above the threshold.
* Find local maxima to obtain the note onsets. 
* Place them in a time series signal. 

Specifications of the windows and filter-kernals were chosen by hand.

The postulated justification for these values would be according to the following:

Let, the smoothing-window's length be $L$ and the hop-size of the analysis window of the FFT be $H$. Since the OSS signal is decimated in time by a factor equal to the hop-size, the length of the smoothing window with respect to the audio signal's sample rate $F_s$ would be:

$L_{n} = H * L$ 

This is in terms of samples. In terms of time, it is equal to, 

$L_{t} = \dfrac{L_{n}}{F_s} $

Since the FFT can capture a transient that is only shorter than the FFT length $N_{FFT}$, it would make sense to set $L_t$ such that,

$L_t < \dfrac{N_{FFT}}{F_s}$

In this case, $N_{FFT}=1024$, $F_s=44100$ therefore, $L_t$ must be less than `23ms`. Back calculating, this yields, L_{n}

```python
def peak_pick(oss, draw_plots=False):
    # Smooth the OSS
    smooth_oss = np.convolve(oss, np.hanning(64))
    
    # Create adaptive threhsold
    adaptive_thresh = medfilt(smooth_oss, 55)

    # Threshold the smoothed-OSS
    thresholded_oss = np.subtract(smooth_oss, adaptive_thresh)
    thresholded_oss = half_wave_rectify(thresholded_oss)
    
    # Find local maxima
    peak_indices = argrelextrema(thresholded_oss, np.greater)
    
    # Generate output signal
    peaks = np.zeros(len(thresholded_oss))
    peaks[peak_indices] = 1.0
    
    # Draw plots to illustrate steps (if required)
    if(draw_plots):
        figure(542)
        subplot(4,1,1), plot(oss, label='OSS'), legend()
        subplot(4,1,2), plot(adaptive_thresh, label='adaptive'), legend()
        subplot(4,1,3), plot(thresholded_oss, label='thresh') , legend()
        subplot(4,1,4), plot(peaks, label='peaks'), legend()
    return peaks

onsets = peak_pick(oss, draw_plots=True)
```
