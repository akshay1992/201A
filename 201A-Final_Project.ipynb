##Step : Onset detection

This method of onset detection uses the $L_2$ Norm on the rectified difference in each spectral bin as described in [1] and some improvements on this method as suggested by [2].

It involves the following steps:

* Compute the STFT with N=1024, Hanning, Hop Size=256
* Compute the difference in the magnitude of each bin with that of the previous analysis frame.
* Rectify the difference values using the following function to identify positive change.
    
    $H(x) = (x + |x|)/2 $ 
    
    This evaluates to zero for all negative values of $x$.
* Accumulate all these rectified values over each window. This yields a 1D signal called the Onset Strength Signal (OSS)

-----

[1] A Tutorial on Onset Detection in Music Signals, J. P. Bello L. Daudet, S Abdallah C Duxbury M Davies M B.Sandler

[2] Michael Krzyzaniak, ASU, mkrzyzan@asu.edu

The onset detection function for the above discussed procedure. 

```python
def detectOnsets(audioFile, NFFT=1024, hop=256, draw_plots=False):
    spgram = generateSpecgram(audioFile, np.hanning, NFFT, hop)
    specDiff = np.diff(abs(spgram))
    rectified = half_wave_rectify(specDiff)
    accum = np.sum(rectified, axis=0)
    
    # Draw plots to illustrate algorithm (if required)
    if(draw_plots):
        subplot(2,1,1), plot(audioFile.audio, label='Audio Signal'), legend()
        subplot(2,1,2), plot(accum, label='OSS'), legend()
        
    return accum
```

##Step : Peak Picking

This is used to extract the exact locations of the peaks in the audio signal using a peak picking algorithm as described below. The following series of steps will attempt to illustrate the algorithm. This more or less follows some of the suggestions given in [1].

Algorithm:

* Smooth the OSS with a Hanning window. `(Window length: 64-points)`
* Make an adaptive threshold signal using a median filter on the smoothed-OSS. `(Kernel-size: 55)`
* Subtract this from the smoothed-OSS.
* Half-wave-rectify to isolate the signal that is above the threshold.
* Find local maxima to obtain the note onsets. 
* Place them in a time series signal. 

Specifications of the windows and filter-kernals were chosen by hand.

```python
def peak_pick(oss, win_len=64, medfilt_kernel_size=55, draw_plots=False):
    # Smooth the OSS
    smooth_oss = np.convolve(oss, np.hanning(win_len))
    
    # Create adaptive threhsold
    adaptive_thresh = scp.medfilt(smooth_oss, medfilt_kernel_size)

    # Threshold the smoothed-OSS
    thresholded_oss = np.subtract(smooth_oss, adaptive_thresh)
    thresholded_oss = half_wave_rectify(thresholded_oss)
    
    # Find local maxima
    peak_indices = scp.argrelextrema(thresholded_oss, np.greater)
    
    # Generate output signal
    peaks = np.zeros(len(thresholded_oss))
    peaks[peak_indices] = thresholded_oss[peak_indices]
    peaks = np.divide(peaks, np.max(peaks))
    
    # Draw plots to illustrate steps (if required)
    if(draw_plots):
        figure(542)
        subplot(4,1,1), plot(oss, label='OSS'), legend()
        subplot(4,1,2), plot(adaptive_thresh, label='adaptive'), legend()
        subplot(4,1,3), plot(thresholded_oss, label='thresh') , legend()
        subplot(4,1,4), plot(peaks, label='peaks'), legend()
    return peaks
```

##Step : Finding Inter-Onset Intervals

The following function finds all the inter-onset intervals (IOI) as defined in . It also has parameters to specify the following:

* The minimum temporal distance between two onsets for the IOI to be considered. 
* The maximum number of onsets allowed in between the two onsets under consideration.

This returns a list of IOIs which can later be used to estimate the tatum of the audio sample. 

--------------------------------------------------------_

[3] Uhle and Herre, "Estimation of Tempo, Micro-time and Time Signature from Percussive Music", 2003 

[4] Seppanen, J. "Tatum grid analysis of musical signals", 2001

```python
def findIOIs(onsets, threshold=10000, max_inbetween=100):
    ioi = []
    onset_times = np.nonzero(onsets)[0]
    for i in range(len(onset_times)):
        j = i+1
        while j<len(onset_times):
            if j-i > max_inbetween+1:
                j = j+1
                break
            diff = onset_times[j]-onset_times[i]
            if diff < threshold:
                 ioi.append(diff) 
            j = j+1
    return ioi
```

In order to test this, we first make an educated guess as to what the theshold should be. 

Remember, that the threshold is in samples of the onset signal; and also that the onsets are obtained from an FFT that had a hop size of 256. Thus we would need to convert the time threshold according to:

$ Thresh_{samples} = Thresh_{seconds} * Sample Rate / 256 $

Assuming that the tatum cannot be larger than 8 seconds (which corresponds to about 60bpm in a 4-beat measure), we can compute the Threshold using the above formula. 

For the sake of testing, we can assume that the max_inbetween for the algorithm to consider can be less than 8. This corresponds to an eigth-note tatum for a 4-beat measure.

```python
Thresh = 8*44100.0/256
max_inbetween = 8

# Using these to compute the IOIs
ioi = findIOIs(onsets, Thresh, max_inbetween)
```

As we can see we can find a large range of values in this. However, it is easy to see that most of the numbers lie within a certain ranges for example, 85-89, and the rest are just multiples of these numbers (oweing to the method used to obtain the IOIs). The following procedures will exploit this fact.

We will now define a function that counts the occurance of each unique value in the IOI data. This is essentially a histogram of the IOI data. 

Additionally, when the 'collect' feature of the function is enabled, it will bin all the values that are near to each other based on the `bin_size`; for example the set ${ 85, 86, 87, 88, 89 }$ will fall into the same bin, with their frequency values accumulated, and the bin value replaced by the mean of ${ 85, 86, 87, 88, 89 }$.

```python
def ioi_hist(ioi, collect=True, bin_size=5):
    y = np.bincount(ioi)
    ii = y.nonzero()[0]
    pre_histogram = np.c_[ii, y[ii]]
    
    if not collect:
        # This is to view and decide the bin size.
        return pre_histogram

    diff = np.diff(ii)
    bin_changes = [0]
    for index in range(len(diff)):
        if( diff[index] > bin_size):
            bin_changes.append( index + 1 )

    bins = []
    for bin_index in range(len(bin_changes)-1):
        this_bin = pre_histogram[ bin_changes[bin_index] : bin_changes[bin_index+1] ]
        accumulate = np.sum( this_bin, axis=0)
        accumulate[0] = accumulate[0] / ( bin_changes[bin_index+1] - bin_changes[bin_index] )
        bins.append(accumulate)

    this_bin = pre_histogram[ bin_changes[-1] : ]
    accumulate = np.sum( this_bin, axis=0)
    accumulate[0] = accumulate[0] / ( len(pre_histogram) - bin_changes[-1] )
    bins.append(accumulate)

    return np.array(bins)
```

```python
IOI_histogram = ioi_hist(ioi)
```

If we plot this as a graph, the tatum is evident: it is the largest peak in the IOI histogram.

In this specific case, it is easy to discern because of the simplicity of the sound sample. All the onsets occur at the same interval and there is no rhythmic complexity. We see the interval with with the lowest period having the highest probability. All the remaining are the multiples of this interval and they have a linearly decreasing probabilities.

In this case the tatum is much harder to discern. The peak at 255 may or may not be the tatum without a priori knowledge of the drum sample. However, we know from the onset signal that the tatum in this case might be the Eighth-notes, since that is the smallest interval present in the signal, this should correspond to the IOI value of around 42 in the signal. But, we see see the prominent peaks at 85, 170, 256 and in-between that are either equal-to, or larger than the peak at 42. 

According to [3] and [4], it seems useful to smooth this histogram with a Gaussian Probaility Density Function. However, given the resolution of this histogram, using a smoothing function does more harm than good. 

The good news however, is that one can already start to see the interval-ratio patterns are starting to show in the histogram. The first three prominent peaks in the above histogram are at: 
* 85
* 170
* 256

Doing some simple integer division, one can obtain some discernable ratios:

$ \frac{256}{85} \approx 3 $ and $ \frac{170}{85} \approx 2 $.

With this alone we can postulate that the metre is a kind of triple metre.
