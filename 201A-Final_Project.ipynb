##Step 1: Onset Strength Signal

This method of onset detection uses the rectified difference of the spectral flux as described in [1] and [2].

It involves the following steps:

* Compute the STFT with N=1024, Hanning, Hop Size=256
* Compute the difference in the magnitude of each bin with that of the previous analysis frame. (Spectral Flux)
* Half-wave rectify the difference values using the following function to identify positive change.
    
    $H(x) = \frac{(x + |x|)}{2} $ 
    
    This evaluates to zero for all negative values of $x$.
* Accumulate all these rectified values over each window. This yields a 1D signal called the Onset Strength Signal (OSS)

-----

[1] Duxbury et al, "A Tutorial on Onset Detection in Music Signals", 2005

[2] Scheirer and Slaney, "Construction and Evaluation of a Robust MultiFeature Speech-Music Discriminator", 1997

```python
a = audioFile('4-4.wav') 
a.downmixR()

plot(a.audio)
```

The onset detection function for the above discussed procedure. 

```python
def detectOnsets(audioFile, NFFT=1024, hop=256, draw_plots=False):
    spgram = generateSpecgram(audioFile, np.hanning, NFFT, hop)
    specDiff = np.diff(abs(spgram))
    rectified = half_wave_rectify(specDiff)
    accum = np.sum(rectified, axis=0)
    
    # Draw plots to illustrate algorithm (if required)
    if(draw_plots):
        subplot(2,1,1), plot(audioFile.audio, label='Audio Signal'), legend()
        xlim(0, len(audioFile.audio))
        subplot(2,1,2), plot(accum, label='OSS'), legend()
        xlim(0, len(accum))
        
    return accum
```

```python
oss = detectOnsets(a, draw_plots=True)
```

It is easy to see that the onsets are clearly extracted from this signal. This basic algorithm will be used for the rest of the project. 

##Step 2: Peak Picking

This is used to extract the exact locations of the peaks in the audio signal using a peak picking algorithm as described below. The following series of steps will attempt to illustrate the algorithm. This more or less follows some of the suggestions given in [1].

Algorithm:

* Smooth the OSS with a Hanning window. 
* Make an adaptive threshold signal using a median filter on the smoothed-OSS.
* Subtract this from the smoothed-OSS.
* Half-wave-rectify to isolate the signal that is above the threshold.
* Find local maxima to obtain the note onsets. 
* Place them in a time series signal. 

The following specifications of the windows and filter-kernals were chosen by hand.

    Window length: 64
    Kernel-size: 55

```python
def peak_pick(oss, win_len=64, medfilt_kernel_size=55, draw_plots=False):
    # Smooth the OSS
    smooth_oss = np.convolve(oss, np.hanning(win_len))
    
    # Create adaptive threhsold
    adaptive_thresh = scp.medfilt(smooth_oss, medfilt_kernel_size)

    # Threshold the smoothed-OSS
    thresholded_oss = np.subtract(smooth_oss, adaptive_thresh)
    thresholded_oss = half_wave_rectify(thresholded_oss)
    
    # Find local maxima
    peak_indices = scp.argrelextrema(thresholded_oss, np.greater)
    
    # Generate output signal
    peaks = np.zeros(len(thresholded_oss))
    peaks[peak_indices] = thresholded_oss[peak_indices]
    peaks = np.divide(peaks, np.max(peaks))
    
    # Draw plots to illustrate steps (if required)
    if(draw_plots):
        figure(542)
        subplot(4,1,1), plot(oss, label='OSS'), legend()
        subplot(4,1,2), plot(adaptive_thresh, label='adaptive'), legend()
        subplot(4,1,3), plot(thresholded_oss, label='thresh') , legend()
        subplot(4,1,4), plot(peaks, label='peaks'), legend()
    return peaks
```

##Step 3: Finding Inter-Onset Intervals

The following function finds all the inter-onset intervals (IOI) as defined by Dixon [3]. This takes into account the time differences between not only the successive onsets, but all onsets within a specified duration. It also has parameters to specify the following:

* The minimum temporal distance between two onsets for the IOI to be considered. 
* The maximum number of onsets allowed in between the two onsets under consideration.

This returns a list of IOIs which can later be used to estimate the tatum [4] of the audio sample. 

--------------------------------------------------------

[3] Simon E. Dixon, “A beat tracking system for audio signals”, 1999

[4] Jeffrey A. Bilmes, "Timing Is Of Essence", 1993

```python
def findIOIs(onsets, threshold=10000, max_inbetween=100):
    ioi = []
    onset_times = np.nonzero(onsets)[0]
    for i in range(len(onset_times)):
        j = i+1
        while j<len(onset_times):
            if j-i > max_inbetween+1:
                j = j+1
                break
            diff = onset_times[j]-onset_times[i]
            if diff < threshold:
                 ioi.append(diff) 
            j = j+1
    return ioi
```

In order to test this, we first make an educated guess as to what the theshold should be. 

Remember, that the threshold is in samples of the onset signal; and also that the onsets are obtained from an FFT that had a hop size of 256. Thus we would need to convert the time threshold according to:

$ Thresh_{samples} = \frac{Thresh_{seconds} * Sample Rate}{ 256 }$

Assuming that the tatum cannot be larger than 8 seconds (which corresponds to about 60bpm in a 4-beat measure), we can compute the Threshold using the above formula. 

For the sake of testing, we can assume that the `max_inbetween` for the algorithm to consider can be less than 8. This corresponds to an eigth-note tatum for a 4-beat measure.

```python
Thresh = 8*44100.0/256
max_inbetween = 8

# Using these to compute the IOIs
ioi = findIOIs(onsets, Thresh, max_inbetween)
```

##Step 4: Tatum Estimation using an IOI Histogram

We will now define a function that counts the occurance of each unique value in the IOI data. This is essentially a histogram of the IOI data. 

Additionally, when the 'collect' feature of the function is enabled, it will bin all the values that are near to each other based on the `bin_size`; for example the set ${ 85, 86, 87, 88, 89 }$ will fall into the same bin, with their frequency values accumulated, and the bin value replaced by the mean of ${ 85, 86, 87, 88, 89 }$.

```python
def ioi_hist(ioi, collect=True, bin_size=5):
    y = np.bincount(ioi)
    ii = y.nonzero()[0]
    pre_histogram = np.c_[ii, y[ii]]
    
    if not collect:
        # This is to view and decide the bin size.
        return pre_histogram

    diff = np.diff(ii)
    bin_changes = [0]
    for index in range(len(diff)):
        if( diff[index] > bin_size):
            bin_changes.append( index + 1 )

    bins = []
    for bin_index in range(len(bin_changes)-1):
        this_bin = pre_histogram[ bin_changes[bin_index] : bin_changes[bin_index+1] ]
        accumulate = np.sum( this_bin, axis=0)
        accumulate[0] = accumulate[0] / ( bin_changes[bin_index+1] - bin_changes[bin_index] )
        bins.append(accumulate)

    this_bin = pre_histogram[ bin_changes[-1] : ]
    accumulate = np.sum( this_bin, axis=0)
    accumulate[0] = accumulate[0] / ( len(pre_histogram) - bin_changes[-1] )
    bins.append(accumulate)

    return np.array(bins)
```

## Step 7: Ratio-Matrix

This section tries to develop the idea of a Ratio-Matrix that is an abstraction of the IOI ratios in the given space. 


The first step in developing the Ratio-Matrix is to obtain an ordered set of IOIs. The following function tries to obtain the list of IOIs in the descending order of frequencies of occurance. This is done based on the ACF defined above. It does so by the following process: 

* Smooth the ACF with a Hamming window
* Use a peak-picking algorithm to obtain the peaks 
* Sort the peaks in descending order
* Output a matrix with columns `[IOI; ACF-Value]`

```python
def orderedIOIs( autocorr ):
    autocorr = np.convolve(autocorr, np.hamming(32), mode='same' )
    peaks = scp.argrelextrema(acor2, np.greater)[0]
    values = autocorr[peaks]
    mat = np.c_[peaks, values]
    
    mat = mat[mat[:,1].argsort(axis=0)] # Sort
    mat = mat[::-1]                     # Reverse Order
    return mat
```

```python
ioi2_ordered = orderedIOIs(acor2)
print ioi2_ordered
```

 Once this is obtained, it is worth investigating the way the frequency of the IOIs decrease over this list. 

```python
plot(ioi2_ordered[:,1])
show()
```

The above plot shows that it is merely a linear decrease. Therefore we could simply take the first N highest IOIs in order to form some sort of a heuristic to guess the tempo. This is investigated using the Ratio-Matrix.

The Ratio-Matrix is defined and obtained as folllows. The IOIs are first ordered with a decreasing order of frequencies. Then, the top N candidates from this list are taken as a vector $V$, where $V$ is a row vector having dimension $1\times N$. Then ratio-matrix $M_R$ is obtained using: 

$M_R = H(V)^T * V $

Where H(A) is defined as 

$H(A) = [\frac{1}{a}, \frac{1}{b} , \frac{1}{c}] $ for $A = [a, b, c]$

This results in a matrix of dimensions $ N \times N $. Once this matrix is obtained, the values within the matrix are rounded to an integer to obtain whole number ratios. 

```python
def ioi_ratios(ioi_ordered):
    ioi_vector = np.matrix(ioi_ordered)
    ratio_matrix = np.dot( (1.0/ioi_vector).T, ioi_vector )
    ratio_matrix = np.round(ratio_matrix)
    print ratio_matrix
```

```python
top_ioi2 = ioi2_ordered[:5,0]
ioi_ratios(top_ioi2)
```

The above ratio-matrix must be read in a specific manner. The top left corner of the image is the most frequent interval. And the frequency decreases both row-wise and column-wise as we go further down to the bottom right. Therefore, the top left corner can be read as the corner with the most-likely metre and the bottom left as the least-likely. 

Furthermore, we could weed out the values $ 86, 29, 58, 43 $ and all the $1$s and $0$s with the notion that they are unlikely metres. We then get the whole number ratios 3 and 2 as the most likely candidates. 

It is possible to suggest that this might be the first step toward a guess at a time-signature. However this ought to be tested on other metres.
