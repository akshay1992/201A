###Generalizing

```python
# A function for extracting the data. 
def extract_column(country_name, csv_file):
    f = open(csv_file)
    reader = csv.reader(f)

    x_data = []
    y_data = []

    
    for row in reader:
        if 'Oil' in row[0]:
            x_data = row
        if country_name in row[0]:
            y_data = row
            
    if len(y_data) == 0:
        print "Data not found"
        return 0, 0
    else:
        for word in y_data:
            if word == '-':
                print 'Missing data'
                return 0,0
        
    f.close()
    return x_data, y_data
```

```python
# A function to plot the data
def custom_plot(x_data, y_data, stable_axis=True):
    if x_data == 0 and y_data == 0:
        return
    
    # Remove all non-numerical elements (Missing data for year 2011)
    x = x_data[1:-1]
    y = y_data[1:-1]
    
    # Convert strings into floats
    x = [float(value) for value in x]
    y = [float(value) for value in y]
    
    # Plot data
    plot(x, y)
    ylabel(x_data[0])
    xlabel('Year')
    title(y_data[0])
    show()
```
